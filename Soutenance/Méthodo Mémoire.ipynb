# Méthodologie Économétrique Détaillée  
*Étude de l’incertitude économique et de la tarification des actions*

---

## 1. Construction et choix de la mesure d’incertitude économique

- **Indicateur utilisé** :  
  Indice d’incertitude macroéconomique de Jurado, Ludvigson et Ng (JLN), reconnu pour sa robustesse et largement utilisé dans la littérature.
- **Principe** :
  - Agrège l’imprévisibilité (volatilité inattendue) de nombreuses séries macroéconomiques US (production, emploi, inflation, etc.).
  - À chaque mois, calcule la part d’incertitude non expliquée par les modèles autorégressifs pour chaque série, puis agrège ces innovations pour obtenir un indice synthétique.
  - Différents horizons existent (1, 3, 12 mois), mais l’horizon 1 mois est le plus couramment utilisé.

---

## 2. Estimation du beta d’incertitude (βUNC) pour chaque action

- **Objectif** :  
  Quantifier, à chaque date, la sensibilité du rendement de chaque action à l’incertitude macroéconomique.
- **Procédure** :
  1. **Fenêtre glissante** :
      - Pour chaque action *i* et chaque mois *t*, on utilise les 60 derniers mois de données (de *t-59* à *t*).
      - Cela permet d’obtenir un estimateur dynamique, actualisé à chaque période.
  2. **Régression de référence** :
      $$
      R_{i, \tau} - R_{f, \tau} = \alpha_{i, t} + \beta_{UNC, i, t} \cdot UNC_\tau + \beta_{MKT, i, t} \cdot MKT_\tau + \epsilon_{i, \tau}
      $$
      - $R_{i, \tau}$ : rendement de l’action $i$ au mois $\tau$
      - $R_{f, \tau}$ : taux sans risque au mois $\tau$
      - $UNC_\tau$ : valeur de l’indice d’incertitude JLN au mois $\tau$
      - $MKT_\tau$ : rendement excédentaire du marché
      - Estimation par moindres carrés ordinaires (OLS)
  3. **Résultat** :
      - Chaque action obtient, à chaque date, une valeur de $\beta_{UNC}$ actualisée.

---

## 3. Formation des portefeuilles selon le beta d’incertitude

- **Classement** :
  - À chaque mois, les actions sont triées selon leur $\beta_{UNC}$ estimé.
- **Découpage** :
  - Constitution de 10 portefeuilles (déciles), du plus faible au plus fort $\beta_{UNC}$.
  - Deux types de portefeuilles :
    - **Égal-pondéré** : chaque action a le même poids.
    - **Valeur-pondéré** : pondération par la capitalisation boursière.
- **But** :
  - Comparer les rendements et alphas des extrêmes (décile 1 vs décile 10) pour identifier l’existence d’une prime d’incertitude.

---

## 4. Calcul et analyse des rendements et alphas des portefeuilles

- **Rendement excédentaire** :
  - Calcul du rendement excédentaire moyen pour chaque portefeuille.
- **Alpha factoriel** :
  - Estimation de l’alpha (rendement anormal) par régression sur différents modèles :
    - CAPM (facteur marché)
    - Fama-French 5 facteurs (marché, taille, value, momentum, liquidité)
    - Fama-French-Carhart 7 facteurs (ajout d’investissement et rentabilité)
  - L’alpha est l’ordonnée à l’origine de la régression.
- **Correction des erreurs standards** :
  - Utilisation de Newey–West pour robustesse (autocorrélation et hétéroscédasticité).
- **Interprétation** :
  - Un alpha significatif suggère une surperformance ou sous-performance non expliquée par les facteurs connus.

---

## 5. Analyses conditionnelles par tris bivariés

- **But** :
  - Vérifier si l’effet de $\beta_{UNC}$ subsiste lorsqu’on contrôle d’autres caractéristiques prédictives.
- **Procédure** :
  1. **Tri primaire** : découpage selon une caractéristique (par ex. liquidité : ILLIQ, taille, value…).
  2. **Tri secondaire** : dans chaque groupe primaire, nouveau tri en déciles selon $\beta_{UNC}$.
  3. **Calcul** : rendement/alpha moyen dans chaque sous-groupe.
- **Interprétation** :
  - Si la relation entre $\beta_{UNC}$ et le rendement reste significative dans chaque sous-groupe, cela indique un effet marginal robuste.

---

## 6. Estimation de régressions cross-sectionnelles Fama–MacBeth

- **Objectif** :
  - Mesurer l’effet marginal de $\beta_{UNC}$ sur les rendements futurs, en contrôlant les autres caractéristiques firm-level.
- **Procédure** :
  1. **Pour chaque mois $t$** :
      - Régression cross-sectionnelle :
        $$
        y_{i, t+1} = \gamma_{0, t} + \gamma_{1, t} \cdot \beta_{UNC, i, t} + \sum_k \gamma_{k+1, t} X_{k, i, t} + \varepsilon_{i, t+1}
        $$
        - $y_{i, t+1}$ : rendement (ou alpha) futur de l’action $i$
        - $X_{k, i, t}$ : autres caractéristiques firm-level (taille, value, momentum, illiquidité…)
  2. **Stockage des coefficients** : collecte de $\gamma_{1, t}$ chaque mois pour analyser la moyenne temporelle.
  3. **Significativité** : test de la moyenne et de la t-statistique corrigée Newey-West.
- **Contrôle sectoriel** :
  - Ajout de variables muettes pour chaque secteur afin de contrôler les effets propres à chaque industrie.
- **Interprétation** :
  - Un coefficient moyen négatif et significatif de $\beta_{UNC}$ confirme l’existence d’une prime refuge.

---

## 7. Tests de robustesse et contrôles supplémentaires

- **Variantes** :
  - Différents horizons de prévision (1, 3, 6, 12 mois).
  - Spécifications alternatives (modèles, contrôles sectoriels…).
  - Analyses par sous-périodes (ex : avant/après crise 2008 ou Covid).
- **Gestion des valeurs manquantes** :
  - Nettoyage ciblé : élimination uniquement des observations manquantes nécessaires à chaque régression.
  - Moyennes transversales pour la construction de certains facteurs (ex : illiquidité).
- **But** :  
  - Vérifier la stabilité et la robustesse des résultats obtenus.

---

## 8. Synthèse de l’approche

**Résumé visuel du déroulement** :

1. **Construction indice d’incertitude JLN**
2. **Estimation dynamique de $\beta_{UNC}$ (régression glissante 60 mois)**
3. **Tri en portefeuilles (déciles)**
4. **Calcul rendements/alphas + correction Newey-West**
5. **Tris bivariés conditionnels**
6. **Régressions Fama–MacBeth**
7. **Tests de robustesse (horizons, sous-périodes, spécifications)**
8. **Synthèse des résultats et interprétation économique**

---

## Conseils pour l’explication orale

- Pour chaque étape : expliquer le **but**, la **procédure** concrète, et **l’interprétation économique**.
- Utiliser des exemples concrets (ex : "pour chaque mois, on regarde la sensibilité de chaque action à l’incertitude, puis on voit si cela explique son rendement futur").
- Schématiser la séquence pour faciliter la compréhension.

# Fiche de vocabulaire – Régressions Fama–MacBeth et analyse empirique

---

### 1. **Fama–MacBeth Regression**
**Définition :**  
Méthode d’estimation en deux étapes utilisée pour analyser la relation entre les rendements d’actifs et des caractéristiques firm-level.  
**Explication soutenance :**  
Chaque mois, on fait une régression cross-sectionnelle, puis on moyenne les coefficients sur toute la période afin d’estimer les effets moyens et leur significativité.

---

### 2. **Cross-sectional regression (régression en coupe transversale)**
**Définition :**  
Régression effectuée à un instant donné sur un ensemble d’actions différentes.  
**Explication soutenance :**  
Permet de comprendre quelles variables expliquent les différences de rendements entre entreprises à une date donnée.

---

### 3. **βUNC (Bêta d’incertitude)**
**Définition :**  
Mesure de la sensibilité du rendement d’une action à un facteur d’incertitude macroéconomique.  
**Explication soutenance :**  
Plus ce bêta est élevé, plus l’action réagit aux fluctuations de l’incertitude globale.

---

### 4. **Prime refuge (safe haven premium)**
**Définition :**  
Phénomène où des actifs exposés à l’incertitude affichent des rendements plus faibles, car ils servent de refuge lors des chocs.  
**Explication soutenance :**  
Un coefficient négatif de βUNC indique un effet refuge : les investisseurs acceptent un rendement moindre en échange d’une protection contre l’incertitude.

---

### 5. **Variable dépendante / Variable explicative**
**Définition :**  
- Dépendante : la variable qu’on cherche à expliquer (ici, rendement ou βUNC)  
- Explicative : variable(s) utilisée(s) pour expliquer la variable dépendante (taille, BM, βUNC, etc.)  
**Explication soutenance :**  
On regresse généralement le rendement futur sur des caractéristiques d’entreprise pour voir si celles-ci prédisent la performance.

---

### 6. **Effet de contrôle d’industrie (Industry fixed effects / contrôle sectoriel)**
**Définition :**  
Ajout de variables muettes ("dummies") pour neutraliser l’effet spécifique de chaque secteur dans une régression.  
**Explication soutenance :**  
Permet de comparer les entreprises à l’intérieur d’un même secteur, et de s’assurer que les résultats ne sont pas dus à une sur-représentation d’un secteur.

---

### 7. **Variables de contrôle (Control variables)**
**Définition :**  
Autres caractéristiques prises en compte dans la régression, pour isoler l’effet d’une variable principale.  
**Explication soutenance :**  
Par exemple : taille, valeur comptable (BM), momentum, volatilité, etc.

---

### 8. **t-statistique**
**Définition :**  
Statistique qui mesure la significativité d’un coefficient estimé dans une régression.  
**Explication soutenance :**  
Un t-stat supérieur à 2 en valeur absolue indique généralement une significativité au seuil de 5%.

---

### 9. **Adj. R² (R carré ajusté)**
**Définition :**  
Part de la variance de la variable dépendante expliquée par le modèle, corrigée du nombre de variables utilisées.  
**Explication soutenance :**  
Permet de juger la qualité explicative du modèle, tout en pénalisant l’ajout de variables inutiles.

---

### 12. **Variables firm-level**
**Définition :**  
Caractéristiques propres à chaque entreprise (taille, BM, rentabilité, liquidité…).  
**Explication soutenance :**  
Elles servent à expliquer les variations de βUNC ou des rendements entre entreprises.

---

### 13. **Effets fixes (Fixed effects)**
**Définition :**  
Variables qui capturent l’effet spécifique à un groupe (industrie, période…), constant dans le temps.  
**Explication soutenance :**  
Utilisé pour contrôler l’influence de facteurs non observés mais communs à un groupe.

---

### 14. **Prime de risque (Risk premium)**
**Définition :**  
Rendement supplémentaire exigé par les investisseurs pour détenir un actif risqué.  
**Explication soutenance :**  
La régression cherche à voir si βUNC est associé à une prime de risque positive ou négative.

---

### 15. **Effet sectoriel / composition sectorielle**
**Définition :**  
Influence du poids de chaque secteur dans un échantillon sur les résultats globaux.  
**Explication soutenance :**  
Un secteur surreprésenté ou atypique peut fausser les relations observées si on ne contrôle pas pour l’industrie.

---

### 16. **Correction Newey-West**
**Définition :**  
Méthode d’ajustement des erreurs standard pour tenir compte de l’autocorrélation et de l’hétéroscédasticité dans les séries temporelles.  
**Explication soutenance :**  
Permet d’obtenir des t-statistiques robustes même si les erreurs ne sont pas indépendantes ou de variance constante, ce qui est courant dans les panels financiers.

---

### 17. **Analyse bivariée**
**Définition :**  
Analyse statistique qui étudie la relation entre deux variables uniquement (ex : rendement et βUNC, sans autres contrôles).  
**Explication soutenance :**  
Permet d’observer l’effet “brut” d’une variable sur une autre, avant d’introduire d’autres facteurs explicatifs.

---

### 18. **Co-skewness (COSKEW)**
**Définition :**  
Mesure de la tendance d’un actif à avoir des rendements extrêmes en même temps que le marché (asymétrie commune).  
**Explication soutenance :**  
Un actif avec une co-skewness élevée aura des performances extrêmes quand le marché est extrême : c’est un facteur de risque supplémentaire à la variance.
