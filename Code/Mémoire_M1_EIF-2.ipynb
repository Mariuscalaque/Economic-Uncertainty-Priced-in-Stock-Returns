## Uniformisation des bases de données et détection de la période commune

```r
# 0) Chargement des librairies ------------------------------------------------
library(readxl)   # Nous utilisons ce package pour importer des données depuis des fichiers Excel.
library(dplyr)    # Nous utilisons dplyr pour manipuler facilement nos jeux de données.

# 1) Lecture des fichiers Excel ------------------------------------------------
# Nous lisons chacun des fichiers Excel contenant les différentes sources de données.
# Pour chaque fichier, nous précisons que la première ligne contient les noms de colonnes
# et nous considérons les cellules vides ("") ou contenant "NA" comme des valeurs manquantes.
Risk_Premium      <- read_excel("SP500_Risk_Premium.xlsx",      col_names = TRUE, na = c("", "NA"))
Q_factors         <- read_excel("Q_Factors.xlsx",               col_names = TRUE, na = c("", "NA"))
MacroUncertainty  <- read_excel("MacroUncertaintyData.xlsx",    col_names = TRUE, na = c("", "NA"))
Illiq             <- read_excel("ILLiq_SP500.xlsx",             col_names = TRUE, na = c("", "NA"))
CoSkew            <- read_excel("CoSkew_60m.xlsx",              col_names = TRUE, na = c("", "NA"))
IdioVol           <- read_excel("IdioVol_60m.xlsx",             col_names = TRUE, na = c("", "NA"))
MAX               <- read_excel("MAX_5days.xlsx",               col_names = TRUE, na = c("", "NA"))

# Ainsi, nous centralisons toutes les données nécessaires pour nos analyses financières.

# 2) Renommage de la première colonne en "date" -------------------------------
# Nous rassemblons tous nos jeux de données dans une liste pour appliquer des modifications en lot.
datasets <- list(
  Risk_Premium, Q_factors, MacroUncertainty,
  Illiq, CoSkew, IdioVol, MAX
)
# Nous parcourons chaque jeu de données pour s'assurer que la première colonne porte bien le nom "date".
for(df in seq_along(datasets)) {
  names(datasets[[df]])[1] <- "date"
}
# Nous replaçons chaque jeu de données modifié dans son objet d'origine.
Risk_Premium <- datasets[[1]]
Q_factors    <- datasets[[2]]
MacroUncertainty <- datasets[[3]]
Illiq        <- datasets[[4]]
CoSkew       <- datasets[[5]]
IdioVol      <- datasets[[6]]
MAX          <- datasets[[7]]

# Cette étape garantit l’harmonisation des colonnes pour faciliter les futures jointures ou comparaisons.

# 3) Conversion du format Excel (jours depuis 1900-01-00) en Date -------------
# Les dates importées depuis Excel sont souvent des entiers représentant un nombre de jours depuis une origine.
# Ici, nous convertissons chaque colonne "date" en un objet Date R, en précisant l’origine du comptage Excel.
for(name in c("Risk_Premium","Q_factors","MacroUncertainty",
              "Illiq","CoSkew","IdioVol","MAX")) {
  df <- get(name)
  df$date <- as.Date(df$date, origin = "1899-12-30") # Nous utilisons l’origine Excel standard.
  assign(name, df)
}

# 4) Affichage des plages de dates --------------------------------------------
# Nous définissons une fonction utilitaire pour afficher la plage de dates de chaque jeu de données.
print_range <- function(df, name){
  cat(
    sprintf("%-17s:", name),
    format(min(df$date, na.rm = TRUE), "%Y-%m-%d"),
    "→",
    format(max(df$date, na.rm = TRUE), "%Y-%m-%d"),
    "\n"
  )
}
# Nous appelons cette fonction pour chaque jeu de données afin de vérifier la couverture temporelle.
print_range(Risk_Premium,     "Risk_Premium")
print_range(Q_factors,        "Q_factors")
print_range(MacroUncertainty, "MacroUncertainty")
print_range(Illiq,            "Illiquidity")
print_range(CoSkew,           "CoSkewness")
print_range(IdioVol,          "Idio-vol")
print_range(MAX,              "MAX (5 days)")

# Cette vérification nous assure que toutes nos données couvrent bien la période attendue.

# 5) Détermination et affichage de la fenêtre commune ------------------------
# Nous cherchons à déterminer la période commune à tous les jeux de données,
# c’est-à-dire l’intervalle où chaque source dispose de données disponibles.
start_date <- max(
  min(Risk_Premium$date,      na.rm = TRUE),
  min(Q_factors$date,         na.rm = TRUE),
  min(MacroUncertainty$date,  na.rm = TRUE),
  min(Illiq$date,             na.rm = TRUE),
  min(CoSkew$date,            na.rm = TRUE),
  min(IdioVol$date,           na.rm = TRUE),
  min(MAX$date,               na.rm = TRUE)
)
end_date <- min(
  max(Risk_Premium$date,      na.rm = TRUE),
  max(Q_factors$date,         na.rm = TRUE),
  max(MacroUncertainty$date,  na.rm = TRUE),
  max(Illiq$date,             na.rm = TRUE),
  max(CoSkew$date,            na.rm = TRUE),
  max(IdioVol$date,           na.rm = TRUE),
  max(MAX$date,               na.rm = TRUE)
)
cat(
  "Période retenue :",
  format(start_date, "%Y-%m-%d"),
  "→",
  format(end_date,   "%Y-%m-%d"),
  "\n"
)
# Cette étape est cruciale pour garantir la cohérence lors de l’analyse croisée de ces bases de données.

# 6) Filtrage de chaque table sur la fenêtre commune -------------------------
# Nous filtrons chaque jeu de données pour ne garder que les lignes situées dans la fenêtre commune.
Risk_Premium     <- filter(Risk_Premium,      date >= start_date & date <= end_date)
Q_factors        <- filter(Q_factors,         date >= start_date & date <= end_date)
MacroUncertainty <- filter(MacroUncertainty,  date >= start_date & date <= end_date)
Illiq            <- filter(Illiq,             date >= start_date & date <= end_date)
CoSkew           <- filter(CoSkew,            date >= start_date & date <= end_date)
IdioVol          <- filter(IdioVol,           date >= start_date & date <= end_date)
MAX              <- filter(MAX,               date >= start_date & date <= end_date)

# Cette opération nous permet de travailler sur des données parfaitement alignées temporellement.

# 7) Extraction des tickers ---------------------------------------------------
# Nous extrayons la liste des tickers (identifiants d’actifs financiers) à partir de Risk_Premium,
# en enlevant la colonne "date" qui n’est pas un ticker.
tickers   <- colnames(Risk_Premium)[-1]  # Nous retirons la colonne "date" pour ne garder que les tickers.
n_tickers <- length(tickers)              # Nous comptons le nombre de tickers disponibles.
cat("Nombre de tickers    :", n_tickers, "\n")
# Cette information sera précieuse pour la suite de nos analyses, notamment pour les boucles ou les résumés statistiques.
```

## Réplication de la table 1

```r
# ──────────────────────────────────────────────────────────────────────────────
# 1) Importatuion des packages nécessaires
# ──────────────────────────────────────────────────────────────────────────────
library(dplyr)
library(tidyr)
library(readxl)
library(zoo)
library(sandwich)
library(lmtest)
library(purrr)
```

```r
# ──────────────────────────────────────────────────────────────────────────────
# 2) Mettre Risk_Premium en format long et calculer le rendement excédentaire
# ──────────────────────────────────────────────────────────────────────────────

library(dplyr)   # Nous chargeons dplyr pour la manipulation de données.
library(tidyr)   # Nous chargeons tidyr pour la transformation du format des tables.

rp_long <- Risk_Premium %>%
  # 1) Nous nous assurons que toutes les colonnes correspondant aux tickers sont bien au format numérique,
  # en excluant la colonne 'date' de cette conversion.
  mutate(across(-date, as.numeric)) %>%

  # 2) Nous convertissons la colonne 'date' au format Date si ce n'est pas déjà fait.
  mutate(date = as.Date(date)) %>%

  # 3) Nous passons le tableau en format long : chaque ligne correspond à une date, un ticker et un rendement brut.
  pivot_longer(
    cols      = -date,          # Nous sélectionnons toutes les colonnes sauf 'date' à pivoter.
    names_to  = "ticker",       # Nous créons une nouvelle colonne 'ticker' pour le nom du titre.
    values_to = "raw_ret"       # Les valeurs associées sont placées dans la colonne 'raw_ret' (rendement brut).
  ) %>%

  # 4) Nous forçons à nouveau les valeurs de 'raw_ret' en numérique pour éviter tout souci de type.
  mutate(
    raw_ret = as.numeric(raw_ret)
  ) %>%

  # 5) Nous joignons la table Q_factors pour récupérer le taux sans risque (R_F) associé à chaque date,
  # en nous assurant que R_F est aussi en format numérique.
  left_join(
    Q_factors %>%
      select(date, R_F) %>%
      mutate(R_F = as.numeric(R_F)),
    by = "date"
  ) %>%

  # 6) Nous calculons le rendement excédentaire (exret) en soustrayant le taux sans risque du rendement brut.
  mutate(exret = raw_ret - R_F) %>%

  # 7) Nous filtrons pour ne garder que les observations complètes, c'est-à-dire celles où exret n'est pas NA.
  filter(!is.na(exret))

# Nous jetons un rapide coup d'œil aux 10 premières lignes du résultat pour vérifier le format et la cohérence des données.
rp_long %>% head(10)
```

```r
# ──────────────────────────────────────────────────────────────────────────────
# 3) Préparer l’indice d’incertitude avec un décalage d’un mois
# ──────────────────────────────────────────────────────────────────────────────

unc_next <- MacroUncertainty %>%

  # Nous nous assurons que la colonne 'date' est bien au format Date.
  mutate(date = as.Date(date)) %>%

  # Nous trions les données par date croissante, afin de garantir que le décalage se fasse dans le bon ordre temporel.
  arrange(date) %>%

  # Nous créons un nouveau data frame avec :
  # - la date courante
  # - l’indice d’incertitude (UNC_next) décalé d’un mois vers le futur, c’est-à-dire la valeur suivante de la colonne 'h=1'.
  # Cela nous permet d’associer à chaque date la valeur d’incertitude qui sera observée le mois suivant.
  transmute(date, UNC_next = lead(`h=1`, 1))

# Ainsi, nous obtenons un indice d’incertitude "décalé" qui pourra être utilisé dans des analyses de type prédictif,
# où nous cherchons à expliquer ou anticiper la variable d’incertitude du mois suivant en fonction des informations présentes à la date courante.
```

```r
# ──────────────────────────────────────────────────────────────────────────────
# 4) Calculer β^UNC pour chaque action en rolling window de 60 mois
# ──────────────────────────────────────────────────────────────────────────────

betas <- rp_long %>%
  # Nous ajoutons la colonne d'incertitude future (UNC_next) selon la date.
  left_join(unc_next, by = "date") %>%
  filter(!is.na(UNC_next)) %>%
  arrange(ticker, date) %>%
  group_by(ticker) %>%
  # Nous calculons le beta d'incertitude sur une fenêtre glissante de 60 mois.
  mutate(
    beta_unc = rollapplyr(
      data      = cbind(exret, UNC_next),
      width     = 60,
      by.column = FALSE,
      align     = "right",
      fill      = NA_real_,
      FUN       = function(z) {
        if (any(is.na(z))) return(NA_real_)
        coef(lm(z[,1] ~ z[,2]))[2]
      }
    )
  ) %>%
  ungroup() %>%
  select(date, ticker, beta_unc) %>%
  filter(!is.na(beta_unc))

# Nous obtenons ainsi les betas d’incertitude pour chaque action et chaque mois.
```

Pour chaque titre, cette fonction estime en fenêtres mobiles de 60 mois la pente de la régression mensuelle

excess_ret_𝑡 =𝛼+𝛽UNC_{𝑡−1}+𝜀_𝑡

```r
# ──────────────────────────────────────────────────────────────────────────────
# 5) Former les portefeuilles égal‐pondérés décile‐par‐décile
# ──────────────────────────────────────────────────────────────────────────────

portfolio_returns <- betas %>%
  # Nous joignons les betas calculés avec les rendements originaux, en associant chaque action à chaque date.
  inner_join(rp_long, by = c("date","ticker")) %>%

  # Nous groupons par date pour pouvoir créer les déciles à chaque période.
  group_by(date) %>%
  # Nous classons les actions en 10 groupes (déciles) selon leur beta_unc à chaque date.
  mutate(decile = ntile(beta_unc, 10)) %>%
  ungroup() %>%

  # Nous re-groupons par date et par décile pour calculer les statistiques sur chaque portefeuille décile à chaque date.
  group_by(date, decile) %>%
  summarise(
    avg_beta   = mean(beta_unc, na.rm = TRUE),   # Nous calculons le beta_unc moyen du décile.
    avg_return = mean(raw_ret,  na.rm = TRUE),   # Nous calculons le rendement brut moyen du décile.
    exret      = mean(exret,    na.rm = TRUE),   # Nous calculons le rendement excédentaire moyen du décile.
    .groups    = "drop"
  )

# Nous obtenons ainsi, pour chaque date et chaque décile, les performances moyennes des portefeuilles égal-pondérés selon le niveau de sensibilité à l’incertitude.
```

```r
# ──────────────────────────────────────────────────────────────────────────────
# 6) Fonction utilitaire : extraire alpha et t‐stat Newey–West de l’intercept
# ──────────────────────────────────────────────────────────────────────────────
nw_int <- function(model, lag = 12) {
  V  <- NeweyWest(model, lag = lag, prewhite = FALSE)
  ct <- coeftest(model, vcov = V)
  list(
    alpha  = ct["(Intercept)", "Estimate"],
    t_stat = ct["(Intercept)", "t value"]
  )
}
```

```r
# ──────────────────────────────────────────────────────────────────────────────
# 6bis) Fonction utilitaire Safe pour extraire alpha & t‐stat Newey–West
# ──────────────────────────────────────────────────────────────────────────────

safe_nw <- function(mod, lag = 12) {
  tryCatch({
    # Nous calculons la matrice de variance-covariance robuste de Newey-West pour le modèle donné,
    # avec un nombre de lags spécifié (par défaut 12), sans pré-blanchiment.
    V  <- NeweyWest(mod, lag = lag, prewhite = FALSE)
    # Nous extrayons les résultats des coefficients du modèle en utilisant la matrice robuste.
    ct <- coeftest(mod, vcov = V)
    # Nous récupérons l'alpha (constante) et sa statistique t, en prenant la première ligne (Intercept).
    alpha  <- ct[1, "Estimate"]
    t_stat <- ct[1, "t value"]
    # Nous renvoyons alpha et t_stat sous forme de liste.
    list(alpha = alpha, t_stat = t_stat)
  }, error = function(e) {
    # Si une erreur survient (ex : colinéarité, trop peu d'observations), nous renvoyons NA pour alpha et t_stat.
    list(alpha = NA_real_, t_stat = NA_real_)
  })
}

# Cette fonction nous permet d’extraire facilement l’alpha et sa statistique t corrigée de l’autocorrélation,
# tout en gérant les cas où la régression échoue, pour éviter de stopper l’exécution globale d’un script.
```

```r
# ──────────────────────────────────────────────────────────────────────────────
# 7) Calculer les alphas décile‐par‐décile (CAPM, α₅₁, α₅₂, α₄, α₇)
# ──────────────────────────────────────────────────────────────────────────────

alpha_eqw <- portfolio_returns %>%

  # Nous groupons les données par décile pour faire les calculs séparément pour chaque portefeuille.
  group_by(decile) %>%

  # Nous filtrons pour ne garder que les déciles avec au moins 24 observations.
  filter(n() >= 24) %>%

  # Nous appliquons la suite des calculs à chaque groupe (décile).
  group_map(~{
    df <- .x %>%
      # Nous joignons avec les facteurs, pour chaque date.
      left_join(Q_factors2, by = "date") %>%
      # Nous retirons les lignes incomplètes sur les variables nécessaires aux modèles.
      drop_na(exret,
              R_MKT,   # marché
              R_ME,    # taille
              HML,     # value
              MOM,     # momentum
              R_LIQ,   # liquidité
              R_IA,    # investissement
              R_ROE)   # rentabilité

    # Si l'on n'a plus assez d'observations, on renvoie des NA pour ce décile.
    if(nrow(df) < 24) {
      return(tibble(
        decile         = as.character(.y$decile),
        avg_beta       = mean(.x$avg_beta,    na.rm=TRUE),
        avg_return     = mean(.x$avg_return,  na.rm=TRUE),
        exret          = mean(.x$exret,       na.rm=TRUE),
        capm_alpha     = NA_real_, capm_t       = NA_real_,
        alpha_5f_JLN   = NA_real_, t_5f_JLN     = NA_real_,
        alpha_5f_HXZ   = NA_real_, t_5f_HXZ     = NA_real_,
        alpha_4f_HXZ   = NA_real_, t_4f_HXZ     = NA_real_,
        alpha_7f       = NA_real_, t_7f         = NA_real_
      ))
    }

    # Nous estimons les modèles pour chaque portefeuille décile :
    m1     <- lm(exret ~ R_MKT,                data = df)                                # CAPM
    m5_JLN <- lm(exret ~ R_MKT + R_ME + HML + MOM + R_LIQ, data = df)                    # 5 facteurs JLN
    m5_HXZ <- lm(exret ~ R_MKT + R_ME + HML + R_IA + R_ROE, data = df)                   # 5 facteurs HXZ
    m4_HXZ <- lm(exret ~ R_MKT + R_ME + R_IA + R_ROE,       data = df)                   # 4 facteurs HXZ
    m7     <- lm(exret ~ R_MKT + R_ME + HML + MOM + R_LIQ + R_IA + R_ROE, data = df)     # 7 facteurs

    # Nous extrayons alpha et t-stat Newey-West pour chaque modèle.
    nw1     <- safe_nw(m1)
    nw5_JLN <- safe_nw(m5_JLN)
    nw5_HXZ <- safe_nw(m5_HXZ)
    nw4_HXZ <- safe_nw(m4_HXZ)
    nw7     <- safe_nw(m7)

    # Nous enregistrons les résultats pour ce décile.
    tibble(
      decile         = as.character(.y$decile),
      avg_beta       = mean(.x$avg_beta,    na.rm=TRUE),
      avg_return     = mean(.x$avg_return,  na.rm=TRUE),
      exret          = mean(.x$exret,       na.rm=TRUE),

      capm_alpha     = nw1$alpha,    capm_t     = nw1$t_stat,
      alpha_5f_JLN   = nw5_JLN$alpha, t_5f_JLN   = nw5_JLN$t_stat,
      alpha_5f_HXZ   = nw5_HXZ$alpha, t_5f_HXZ   = nw5_HXZ$t_stat,
      alpha_4f_HXZ   = nw4_HXZ$alpha, t_4f_HXZ   = nw4_HXZ$t_stat,
      alpha_7f       = nw7$alpha,    t_7f       = nw7$t_stat
    )
  }) %>%
  # Nous rassemblons tous les résultats des déciles dans un seul tableau.
  bind_rows() %>%
  # Nous classons les résultats dans l'ordre croissant des déciles.
  arrange(as.integer(decile))

# Ce code nous fournit, pour chaque décile, les alphas et t-statistiques associés à différents modèles factoriels,
# afin de comparer la performance ajustée au risque de chaque portefeuille.
```

> **Remarque méthodologique**  
>
> Dans la Table 1, la variable `R_LIQ` ne correspond pas à notre mesure d’illiquidité au niveau de la firme (`ILLIQ`), mais désigne bien le facteur de liquidité de marché au sens de Pastor et Stambaugh (*market-wide liquidity factor*). Cette variable a été téléchargée séparément, conformément à la méthodologie de Pastor–Stambaugh. Autrement dit, `LIQ` doit être comprise comme un facteur agrégé, capturant les variations de liquidité à l’échelle du marché, et non comme une mesure idiosyncratique d’illiquidité d’un titre individuel.

```r
# ──────────────────────────────────────────────────────────────────────────────
# 8) Calcul du spread “10–1” (High minus Low)
# ──────────────────────────────────────────────────────────────────────────────

spread10_1 <- alpha_eqw %>%
  filter(decile %in% c("1", "10")) %>%
  summarise(
    decile           = "10-1",                   # Nom du spread
    avg_beta         = diff(avg_beta),           # Écart du beta moyen (10 - 1)
    avg_return       = diff(avg_return),         # Écart du rendement moyen (10 - 1)
    exret            = diff(exret),              # Écart du rendement excédentaire moyen (10 - 1)
    capm_alpha       = diff(capm_alpha),         # Écart d'alpha CAPM (10 - 1)
    capm_t           = NA_real_,                 # T-stat non calculés ici
    alpha_5f_JLN     = diff(alpha_5f_JLN),
    t_5f_JLN         = NA_real_,
    alpha_5f_HXZ     = diff(alpha_5f_HXZ),
    t_5f_HXZ         = NA_real_,
    alpha_4f_HXZ     = diff(alpha_4f_HXZ),
    t_4f_HXZ         = NA_real_,
    alpha_7f         = diff(alpha_7f),
    t_7f             = NA_real_
  )

# On concatène ce spread au tableau principal des résultats
alpha_results <- bind_rows(alpha_eqw, spread10_1)
```

> **Note sur la gestion des valeurs manquantes dans les jeux de facteurs**
>
> Pour chacun des deux derniers jeux de facteurs, les valeurs manquantes (`NA`) observées s’expliquent par l’étape de fusion des déciles avec l’ensemble des facteurs, suivie d’un nettoyage global via `drop_na(...)`. Ce nettoyage exige que toutes les variables de facteurs soient simultanément non manquantes pour chaque observation (mois, décile). Ainsi, la présence d’un seul `NA` dans un facteur (par exemple dans `R_LIQ` ou `HML`) entraîne la suppression de toute la ligne correspondante.
>
> Cette contrainte peut aboutir à l’élimination complète de certains déciles, en particulier si le nombre d’observations est réduit ou si les données d’un facteur présentent des « trous ».
>
> D’autre part, si après ce nettoyage il reste moins de 24 observations pour un décile donné, nous retournons volontairement un `tibble` rempli de `NA` pour ce groupe, conformément à la règle méthodologique (« au moins 24 mois pour estimer la régression »). Dans ce cas, l’alpha et sa statistique t restent donc `NA`.
>
> **Conclusion :** il faut bien s’assurer qu’il subsiste au moins 24 observations non manquantes pour chaque régression, après le nettoyage, afin de garantir la robustesse de l’estimation.

> **Remarque méthodologique — Nettoyage ciblé des données**
>
> Afin de limiter la perte d’observations et de réduire le nombre de valeurs manquantes (`NA`), nous procédons désormais à un nettoyage ciblé, effectué séparément pour chaque régression. Concrètement, pour chaque modèle factoriel, seules les variables strictement nécessaires à la régression sont utilisées dans l’étape de nettoyage (`drop_na(...)`). Cela permet d’éviter d’éliminer des observations à cause de la présence de `NA` dans des variables non utilisées dans la régression considérée.  
>
> Cette approche plus flexible maximise le nombre d’observations exploitables pour chaque estimation d’alpha et t-stat.

```r
library(dplyr)

# ──────────────────────────────────────────────────────────────────────────────
# 7) Estimation décile‐par‐décile avec nettoyage spécifique à chaque modèle
# ──────────────────────────────────────────────────────────────────────────────

alpha_eqw <- portfolio_returns %>%
  group_by(decile) %>%
  # On garde uniquement les groupes (déciles) avec au moins 24 observations brutes
  filter(n() >= 24) %>%
  # group_map() applique la fonction à chaque groupe (décile) et permet de manipuler .x (les données du groupe) et .y (la clé du groupe)
  group_map(~{
    # Fusionne les rendements du portefeuille avec les facteurs (jointure sur la date)
    base_df <- .x %>%
      left_join(Q_factors2, by = "date")

    # 1) CAPM : nécessite uniquement exret et R_MKT (nettoyage ciblé)
    df1 <- base_df %>% drop_na(exret, R_MKT)
    m1  <- lm(exret ~ R_MKT, data = df1)
    nw1 <- safe_nw(m1)

    # 2) Modèle 5 facteurs "JLN" : exret + R_MKT + R_ME + HML + MOM + R_LIQ
    # Nettoyage ciblé : seules les variables utiles à ce modèle sont exigées non manquantes
    df5a <- base_df %>% drop_na(exret, R_MKT, R_ME, HML, MOM, R_LIQ)
    m5a  <- lm(exret ~ R_MKT + R_ME + HML + MOM + R_LIQ, data = df5a)
    nw5a <- safe_nw(m5a)

    # 3) Modèle 5 facteurs "HXZ" : exret + R_MKT + R_ME + HML + R_IA + R_ROE
    df5b <- base_df %>% drop_na(exret, R_MKT, R_ME, HML, R_IA, R_ROE)
    m5b  <- lm(exret ~ R_MKT + R_ME + HML + R_IA + R_ROE, data = df5b)
    nw5b <- safe_nw(m5b)

    # 4) Modèle 4 facteurs "HXZ" : exret + R_MKT + R_ME + R_IA + R_ROE
    df4  <- base_df %>% drop_na(exret, R_MKT, R_ME, R_IA, R_ROE)
    m4   <- lm(exret ~ R_MKT + R_ME + R_IA + R_ROE, data = df4)
    nw4  <- safe_nw(m4)

    # 5) Modèle 7 facteurs : exret + tous les facteurs précédents
    df7  <- base_df %>% drop_na(exret, R_MKT, R_ME, HML, MOM, R_LIQ, R_IA, R_ROE)
    m7   <- lm(exret ~ R_MKT + R_ME + HML + MOM + R_LIQ + R_IA + R_ROE, data = df7)
    nw7  <- safe_nw(m7)

    # On assemble les résultats dans un tibble (tableau moderne tidyverse)
    # Si une régression ne peut pas être estimée (trop peu de données après nettoyage), safe_nw renvoie NA pour alpha et t-stat
    tibble(
      decile         = as.character(.y$decile), # Décile traité (1 à 10)
      avg_beta       = mean(.x$avg_beta,   na.rm = TRUE), # Moyenne beta du décile
      avg_return     = mean(.x$avg_return, na.rm = TRUE), # Moyenne rendement du décile
      exret          = mean(.x$exret,      na.rm = TRUE), # Moyenne rendement excédentaire

      capm_alpha     = nw1$alpha,   capm_t      = nw1$t_stat,     # Alpha et t-stat CAPM
      alpha_5f_JLN   = nw5a$alpha,  t_5f_JLN    = nw5a$t_stat,    # Alpha et t-stat 5 facteurs JLN
      alpha_5f_HXZ   = nw5b$alpha,  t_5f_HXZ    = nw5b$t_stat,    # Alpha et t-stat 5 facteurs HXZ
      alpha_4f_HXZ   = nw4$alpha,   t_4f_HXZ    = nw4$t_stat,     # Alpha et t-stat 4 facteurs HXZ
      alpha_7f       = nw7$alpha,   t_7f        = nw7$t_stat      # Alpha et t-stat 7 facteurs
    )
  }) %>%
  # On rassemble les résultats de chaque décile dans un seul tableau
  bind_rows() %>%
  # Trie les déciles dans l'ordre croissant (1 à 10)
  arrange(as.integer(decile))


# ──────────────────────────────────────────────────────────────────────────────
# 8) Calcul du spread “10-1” (High minus Low)
# ──────────────────────────────────────────────────────────────────────────────

# On calcule la différence entre le décile 10 et le décile 1 (portefeuille extrême haut - extrême bas)
spread10_1 <- alpha_eqw %>%
  filter(decile %in% c("1","10")) %>%
  summarise(
    decile           = "10-1",                        # Nom de la ligne spread
    avg_beta         = diff(avg_beta),                # Différence beta moyen (10 - 1)
    avg_return       = diff(avg_return),              # Différence rendement moyen (10 - 1)
    exret            = diff(exret),                   # Différence rendement excédentaire (10 - 1)
    capm_alpha       = diff(capm_alpha),  capm_t       = NA_real_, # Pas de t-stat pour le spread
    alpha_5f_JLN     = diff(alpha_5f_JLN), t_5f_JLN    = NA_real_,
    alpha_5f_HXZ     = diff(alpha_5f_HXZ), t_5f_HXZ    = NA_real_,
    alpha_4f_HXZ     = diff(alpha_4f_HXZ), t_4f_HXZ    = NA_real_,
    alpha_7f         = diff(alpha_7f),     t_7f        = NA_real_
  )

# On ajoute la ligne "10-1" au tableau principal
alpha_results <- bind_rows(alpha_eqw, spread10_1)

# ──────────────────────────────────────────────────────────────────────────────
# 9) Affichage final des résultats
# ──────────────────────────────────────────────────────────────────────────────
print(alpha_results)
```

> **Pourquoi observe-t-on encore des `NA` pour les spreads et certains alphas ?**
>
> Le spread “10–1” correspond à la différence entre les déciles 10 et 1 pour chaque statistique de la table (par exemple : alpha, t-stat, rendement, etc.). La fonction `diff()` utilisée pour calculer ce spread renvoie systématiquement `NA` si l’une des deux valeurs (pour le décile 1 ou 10) est déjà `NA`.
>
> Dans votre cas, toutes les régressions des modèles 5f_JLN, 5f_HXZ, 4f_HXZ et 7f ont retourné `NA` pour les déciles 1 et 10 (souvent à cause d’un manque d’observations non manquantes ou d’un nettoyage trop restrictif). Par conséquent, la ligne du spread “10–1” affiche également `NA` pour toutes ces spécifications.
>
> **Conclusion :** Si l’un des deux extrêmes (décile 1 ou 10) n’a pas pu être estimé (alpha ou t-stat manquant), alors le spread n’est pas calculable et reste `NA`.

> **Note méthodologique – Gestion des valeurs manquantes (NA)**
>
> Dans cette version, nous n’utilisons plus `drop_na()` explicitement sur les variables factorielles avant chaque régression.  
>
> Cela signifie que la fonction `lm()` gère directement les valeurs manquantes grâce à son argument interne `na.action = na.omit` (qui est la valeur par défaut).  
>
> **Conséquence :**  
> - Pour chaque appel à `lm()`, seules les lignes complètes (non manquantes pour toutes les variables du modèle) sont conservées automatiquement par `lm()`.  
> - Cela peut simplifier le code, mais il est important de garder à l’esprit que le nombre d’observations utilisées peut alors varier en fonction de chaque modèle et de la présence de NA dans les variables factorielles concernées.  
>
> **Remarque :**  
> - Il reste conseillé de contrôler le nombre d’observations effectives de chaque régression, afin de s’assurer que les estimations sont réalisées sur un nombre suffisant de points (par exemple, au moins 24 mois comme dans votre règle initiale).

> **Transformation alternative : agrégation de la matrice de liquidité en un seul indice mensuel**
>
> Dans cette version, la matrice de liquidité (habituellement composée de plusieurs mesures ou titres) est résumée en un **indice unique** par mois.  
> Cette approche consiste, par exemple, à faire une moyenne (ou une autre statistique agrégée) de toutes les mesures de liquidité disponibles pour chaque mois, afin d’obtenir une seule valeur représentative par période.
>
> **Attention :**  
> Cette méthode diffère de celle utilisée dans l’article
> L’utilisation d’un indice agrégé peut perdre une partie de l’information ou de l’hétérogénéité capturée par l’approche originale.

> **Construction du facteur mensuel d’illiquidité**
>
> Nous avons choisi d’agréger notre mesure firm-level d’illiquidité (ILLIQ) en un facteur mensuel en calculant la **moyenne cross-sectionnelle** de toutes les valeurs d’ILLIQ chaque mois.  
> Concrètement, pour chaque date $t$, nous prenons la moyenne arithmétique des ratios d’illiquidité de tous les titres disponibles, après avoir remplacé les valeurs infinies par des $NA$ et ignoré les titres manquants.  
> Cette approche permet de capturer l’évolution générale de la liquidité du marché tout en restant fidèle à l’idée d’un facteur non échangé (*non–traded liquidity factor*).
>
> Bien que cette méthode ne réplique pas à l’identique la procédure de Pastor–Stambaugh (qui repose sur des portefeuilles 10–1 classés sur des bêta‐de‐liquidité historiques), elle fournit une **approximation robuste** et cohérente avec nos autres sources de données, tout en garantissant une **continuité temporelle** sur notre période d’étude.

```r
library(dplyr)

liquidity_vector <- ILLiq_SP500 %>%
  # 1) Conversion des types : la colonne 'date' en Date, toutes les autres colonnes (tickers) en numérique
  mutate(date = as.Date(date)) %>%
  mutate(across(-date, as.numeric)) %>%

  # 2) Remplacement des valeurs infinies (Inf, -Inf) par NA dans toutes les colonnes de tickers
  mutate(across(-date, ~ ifelse(is.infinite(.), NA, .))) %>%

  # 3) Calcul de la liquidité moyenne mensuelle
  # Pour chaque ligne (date), on calcule la moyenne des valeurs disponibles sur tous les titres, en ignorant les NA
  rowwise() %>%
    mutate(LIQ = mean(c_across(-date), na.rm = TRUE)) %>%
  ungroup() %>%
  select(date, LIQ) # Ne garder que la date et l'indice de liquidité agrégé

# Vérification : on s'assure qu'il ne reste plus de valeurs non numériques dans LIQ
liquidity_vector %>% filter(is.infinite(LIQ) | is.nan(LIQ))
# => Ce filtre doit retourner un tableau vide si tout est correct

# Aperçu des premières et dernières valeurs du vecteur de liquidité
head(liquidity_vector)
tail(liquidity_vector)
```

```r
# ──────────────────────────────────────────────────────────────────────────────
# 1) Calcul des betas d'incertitude macroéconomique (rolling window)
# ──────────────────────────────────────────────────────────────────────────────
library(dplyr)

calculate_uncertainty_betas <- function(stock_names, Risk_Premium, MacroUncertainty) {
  # Préparation du facteur d'incertitude macro : sélection de la colonne UNC et tri par date
  unc_data <- MacroUncertainty %>%
    select(date, UNC = `h=1`) %>%
    arrange(date)

  all_betas <- list() # Initialisation de la liste de résultats individuels

  for (stk in stock_names) {
    message("Calcul des betas pour ", stk)

    # Construction de la table de travail : rendements du titre, jointure sur l'incertitude macro
    cd <- Risk_Premium %>%
      select(date, excess_ret = all_of(stk)) %>%
      arrange(date) %>%
      left_join(unc_data, by = "date") %>%
      mutate(UNC_next = lead(UNC, 1)) # On décale UNC d'un mois (prévisionnelle)

    # Vérification : au moins 60 observations nécessaires pour lancer la fenêtre mobile
    if (nrow(cd) < 60) next

    n <- nrow(cd)
    # Création du tibble résultat pour ce titre : une ligne par date où la régression est possible
    res <- tibble(
      date     = cd$date[60:n],
      stock    = stk,
      beta_unc = NA_real_
    )

    # Fenêtre mobile de 60 mois : à chaque date t, estimation du beta sur les 60 observations précédentes
    for (i in 60:n) {
      wind <- cd[(i-59):i,]
      # On ne lance la régression que si aucune valeur manquante
      if (anyNA(wind$excess_ret) || anyNA(wind$UNC_next)) next
      m <- lm(excess_ret ~ UNC_next, data = wind)
      res$beta_unc[i-59] <- coef(m)[2] # Extraction du coefficient du facteur d'incertitude
    }

    all_betas[[stk]] <- res # Stockage des résultats par titre
  }

  bind_rows(all_betas) # Fusion de tous les résultats individuels dans un seul tibble
}

# Paramètres : noms des titres à traiter (hors colonne "date")
stock_names       <- setdiff(colnames(Risk_Premium), "date")
# Calcul des betas d'incertitude macroéconomique pour tous les titres
uncertainty_betas <- calculate_uncertainty_betas(
  stock_names, Risk_Premium, MacroUncertainty
)
```

```r
# ──────────────────────────────────────────────────────────────────────────────
# 2) Formation des portefeuilles selon les bêtas d'incertitude
# ──────────────────────────────────────────────────────────────────────────────
library(dplyr)
library(tidyr)

form_portfolios <- function(uncertainty_betas, Risk_Premium) {
  # 1) Conversion : toutes les colonnes sauf "date" sont forcées en numérique
  rp_long <- Risk_Premium %>%
    mutate(across(-date, as.numeric)) %>%
    pivot_longer(
      cols        = -date,
      names_to    = "stock",
      values_to   = "ret"
    ) %>%
    # Sécurité : s'assurer que la colonne des rendements est bien numérique
    mutate(ret = as.numeric(ret))

  # 2) Jointure avec les betas d'incertitude et création des déciles
  rp_long %>%
    inner_join(uncertainty_betas, by = c("date","stock")) %>%
    group_by(date) %>%
    # Découpage des titres en 10 groupes (déciles) selon beta_unc à chaque date
    mutate(decile = ntile(beta_unc, 10)) %>%
    group_by(date, decile) %>%
    summarise(
      avg_beta   = mean(beta_unc, na.rm = TRUE), # Beta moyen du décile
      avg_return = mean(ret,      na.rm = TRUE), # Rendement moyen du décile
      n_stocks   = n(),                          # Nombre de titres dans le décile
      .groups    = "drop"
    )
}

# Exemple d’appel : calcule les rendements par décile de beta d'incertitude
portfolio_returns <- form_portfolios(
  uncertainty_betas = uncertainty_betas,
  Risk_Premium      = Risk_Premium
)
```

```r
# ──────────────────────────────────────────────────────────────────────────────
# 3) Calcul des alphas pour chaque décile de portefeuille
# ──────────────────────────────────────────────────────────────────────────────
calculate_alphas <- function(portfolio_returns) {
  # Préparation des facteurs : création d'un jeu de données factoriel avec la colonne date
  factor_data <- Q_factors %>%
    rename(date = date)  # S'assurer que le nom de colonne est correct

  # Initialisation d'un data.frame vide pour stocker les résultats
  alpha_results <- data.frame(
    decile         = character(),   # On stocke les déciles en tant que character
    avg_beta       = double(),
    avg_return     = double(),
    excess_return  = double(),
    capm_alpha     = double(),
    capm_t         = double(),
    capm_beta      = double(),
    q_alpha        = double(),
    q_t            = double(),
    n_months       = integer(),
    stringsAsFactors = FALSE
  )

  for (d in 1:10) {
    # Sélection des données du décile courant, uniquement la date et le rendement moyen
    dec_data <- portfolio_returns %>%
      filter(decile == d) %>%
      select(date, avg_return)

    # Sauter si trop peu de mois disponibles pour le décile
    if (nrow(dec_data) < 24) next

    # Jointure avec les facteurs et calcul du rendement excédentaire sur chaque mois
    cd <- merge(dec_data, factor_data, by = "date")
    cd$excess_return <- cd$avg_return - cd$R_F

    # Estimation du modèle CAPM : alpha, t-stat associé, beta CAPM
    capm_m <- lm(excess_return ~ R_MKT, data = cd)
    capm_a <- coef(capm_m)[1]; capm_t <- summary(capm_m)$coef[1,3]; capm_b <- coef(capm_m)[2]

    # Estimation du modèle Q-factors (Hou, Xue, Zhang) : alpha et t-stat associé
    q_m <- lm(excess_return ~ R_MKT + R_ME + R_IA + R_ROE, data = cd)
    q_a <- coef(q_m)[1]; q_t <- summary(q_m)$coef[1,3]

    # Ajout des résultats pour le décile courant
    alpha_results <- rbind(
      alpha_results,
      data.frame(
        decile         = as.character(d),
        avg_beta       = mean(portfolio_returns$avg_beta[portfolio_returns$decile==d], na.rm=TRUE),
        avg_return     = mean(cd$avg_return, na.rm=TRUE),
        excess_return  = mean(cd$excess_return, na.rm=TRUE),
        capm_alpha     = capm_a,
        capm_t         = capm_t,
        capm_beta      = capm_b,
        q_alpha        = q_a,
        q_t            = q_t,
        n_months       = nrow(cd),
        stringsAsFactors = FALSE
      )
    )
  }

  # Calcul du portefeuille long-short (décile 10 - décile 1)
  if (nrow(alpha_results) == 10) {
    ls <- data.frame(
      decile         = "10-1",
      avg_beta       = alpha_results$avg_beta[10] - alpha_results$avg_beta[1],
      avg_return     = alpha_results$avg_return[10] - alpha_results$avg_return[1],
      excess_return  = alpha_results$excess_return[10] - alpha_results$excess_return[1],
      capm_alpha     = alpha_results$capm_alpha[10] - alpha_results$capm_alpha[1],
      capm_t         = NA,     # t-stat non calculé pour le spread
      capm_beta      = NA,
      q_alpha        = alpha_results$q_alpha[10] - alpha_results$q_alpha[1],
      q_t            = NA,     # t-stat non calculé pour le spread
      n_months       = min(alpha_results$n_months),
      stringsAsFactors = FALSE
    )

    alpha_results <- bind_rows(alpha_results, ls)
  }

  return(alpha_results)
}

# Appel de la fonction pour obtenir les alphas de portefeuille
alpha_results <- calculate_alphas(portfolio_returns)
```

Avec les statistiques Newey-West

```r
# Packages nécessaires
library(dplyr)
library(tidyr)
library(sandwich)   # Pour la correction Newey–West des erreurs
library(lmtest)     # Pour coeftest (t-stats robustes)

# 1) Vérification : la longueur de liquidity_vector doit couvrir la période Q_factors
length_liq <- length(unlist(liquidity_vector))
n_qf       <- nrow(Q_factors)
stopifnot(length_liq >= n_qf)

# 2) Construction de Q_factors2 avec ajout de la colonne LIQ
Q_factors2 <- Q_factors %>%
  mutate(
    LIQ = unlist(liquidity_vector)[ seq_len(n_qf) ]
  )
stopifnot(nrow(Q_factors2) == length(Q_factors2$LIQ))

# 3) Estimation des alphas et des t-stats Newey–West par décile de portefeuille
alpha_eqw <- portfolio_returns %>%
  filter(!is.na(decile)) %>%
  group_by(decile) %>%
  filter(n() >= 24) %>%  # On retient uniquement les déciles suffisamment renseignés
  do({
    df  <- .
    df2 <- df %>%
      left_join(Q_factors2, by = "date") %>%
      mutate(exret = avg_return - R_F) %>%
      drop_na(exret, R_MKT, R_ME, HML, MOM, LIQ, R_IA, R_ROE)

    # Régression CAPM (1 facteur)
    capm5 <- lm(exret ~ R_MKT, data = df2)
    # Régression 5 facteurs (incluant la liquidité)
    ff5   <- lm(exret ~ R_MKT + R_ME + HML + MOM + LIQ,      data = df2)
    # Régression 7 facteurs
    ff7   <- lm(exret ~ R_MKT + R_ME + HML + MOM + LIQ + R_IA + R_ROE,
                data = df2)

    # Covariances Newey–West (lags = 12 pour données mensuelles)
    vcov_capm5 <- NeweyWest(capm5, lag = 12, prewhite = FALSE)
    vcov_ff5   <- NeweyWest(ff5,   lag = 12, prewhite = FALSE)
    vcov_ff7   <- NeweyWest(ff7,   lag = 12, prewhite = FALSE)

    # Extraction des t-stats sur l'intercept (alpha)
    t_capm5 <- coeftest(capm5, vcov = vcov_capm5)["(Intercept)", "t value"]
    t_ff5   <- coeftest(ff5,   vcov = vcov_ff5  )["(Intercept)", "t value"]
    t_ff7   <- coeftest(ff7,   vcov = vcov_ff7  )["(Intercept)", "t value"]

    tibble(
      decile      = as.character(unique(df$decile)),
      avg_beta    = mean(df$avg_beta,    na.rm = TRUE),
      avg_return  = mean(df$avg_return,  na.rm = TRUE),
      exret       = mean(df2$exret,      na.rm = TRUE),
      alpha_CAPM  = coef(capm5)[1],
      t_CAPM      = t_capm5,
      alpha_5fac  = coef(ff5  )[1],
      t_5fac      = t_ff5,
      alpha_7fac  = coef(ff7  )[1],
      t_7fac      = t_ff7
    )
  }) %>%
  ungroup() %>%
  arrange(as.integer(decile))

# 4) Calcul du portefeuille long-short (spread « 10-1 »)
ls10_1 <- alpha_eqw %>%
  filter(decile %in% c("1","10")) %>%
  summarise(
    decile      = "10-1",
    avg_beta    = diff(avg_beta),
    avg_return  = diff(avg_return),
    exret       = diff(exret),
    alpha_CAPM  = diff(alpha_CAPM),
    t_CAPM      = NA_real_,  # Pour le t-stat du spread, il faut le recalculer explicitement si besoin
    alpha_5fac  = diff(alpha_5fac),
    t_5fac      = NA_real_,
    alpha_7fac  = diff(alpha_7fac),
    t_7fac      = NA_real_
  )

# Fusion des résultats décile par décile + spread long-short
alpha_results <- bind_rows(alpha_eqw, ls10_1)

# 5) Affichage du tableau final
print(alpha_results)
```

##Réplication de la Table 2

> **Table 2 : Explication de la méthodologie**
>
> Contrairement à la Table 1, qui s’appuie sur des “factor returns” globaux (facteurs de marché et autres facteurs synthétiques calculés à l’échelle de l’ensemble du marché), la Table 2 est construite dans un univers de **firm-level characteristics**.
>
> Autrement dit, les variables explicatives utilisées dans la Table 2 sont toutes des caractéristiques propres à chaque titre pour chaque mois : il ne s’agit pas de facteurs macro ou de portefeuilles factoriels, mais bien de variables qui décrivent la situation individuelle de chaque action à une date donnée (par exemple : taille, book-to-market, illiquidité individuelle, momentum, etc.).
>
> Ces caractéristiques sont employées dans des régressions en coupe transversale (cross-sectionnelles) pour **prédire la persistance de βUNC** (le beta d’incertitude macroéconomique estimé à l’échelle individuelle pour chaque titre).
>
> Ainsi, la Table 2 met l’accent sur la dimension microéconomique et la dynamique propre à chaque titre, à la différence des analyses purement factorielles de la Table 1.

> **Variables firm-level utilisées dans la Table 2**
>
> La Table 2 mobilise exclusivement des variables **caractéristiques individuelles** (firm-level) pour chaque titre $i$ et chaque mois $t$, qui servent de variables explicatives dans les régressions cross-sectionnelles visant à expliquer la persistance de $\beta^{UNC}$ (beta d’incertitude macroéconomique).  
> Voici la liste des variables et leur mode de calcul, toutes évaluées à la date $t-1$ sauf indication contraire :
>
> - **$\beta^{UNC}_{i,t}$, $\beta^{UNC}_{i,t-1}$** : Beta d’incertitude estimé rolling sur 60 mois via la régression $exret_{i,t} \sim UNC_{t+1}$ (déjà en place).
> - **$\beta^{MKT}_{i,t-1}$** : Beta de marché, rolling 60 mois sur $exret_{i,t} \sim MKT_{t}$ (prime de marché de French).
> - **$SIZE_{i,t-1}$** : Taille, calculée comme $log(\text{MarketCap}) = log(\text{prix} \times \text{actions en circulation})$.
> - **$BM_{i,t-1}$** : Book-to-Market, rapport entre fonds propres comptables et capitalisation boursière.
> - **$REV_{i,t-1}$** : Reversal, rendement du titre au mois $t-1$ : $REV_{i,t-1} = ret_{i,t-1}$.
> - **$MOM_{i,t-1}$** : Momentum, agrégation (somme ou produit) des rendements de $t-12$ à $t-2$ (formule de Carhart, 1997).
> - **$ILLIQ_{i,t-1}$, $COSKEW_{i,t-1}$, $IVOL_{i,t-1}$, $MAX_{i,t-1}$** : Mesures d’illiquidité, coskewness, volatilité idiosyncratique et “lottery demand”, **déjà calculées/importées** au niveau individuel.
> - **$DISP_{i,t-1}$** : Dispersion des prévisions d’analystes, mesurée comme l’écart-type des BPA prévus (sources : IBES/FactSet).
> - **$\beta^{VXO}_{i,t-1}$** : Beta VXO/VIX, rolling 60 mois sur $exret_{i,t} \sim \text{innovation VXO}$.
> - **$R_{IA_{i,t-1}}, R_{ROE_{i,t-1}}$** : Facteurs investissement et profitabilité individuels de Hou–Xue–Zhang (importés via Q_factors).
> - **$MAX_{i,t-1}$** : Proxy “lottery demand”.
>
> Toutes ces variables sont utilisées comme **prédicteurs microéconomiques** dans les régressions en coupe transversale sur la persistance des betas d’incertitude.

# Les données de panel en économétrie financière : définition et utilité

## Qu’est-ce qu’un panel ?
Un **panel** (ou données de panel) désigne un jeu de données **à la fois longitudinales et transversales** :
- **Longitudinales** : on observe plusieurs périodes (mois, trimestres, années, …)
- **Transversales** : on observe plusieurs entités (actions, pays, banques, individus, …)

## Pourquoi utiliser un panel ?

- **Exploiter à la fois la dimension temps et la dimension entité**
    - Suivre l’évolution d’un même titre au fil du temps
    - Comparer simultanément des titres entre eux

- **Contrôler pour l’hétérogénéité inobservable**
    - En régressions de panel à effets fixes, on neutralise tout ce qui est propre à chaque titre (ou à chaque période)

- **Faire des tests cross-sectionnels dynamiques**
    - Exemple : Méthode Fama–MacBeth : chaque mois on fait une régression cross-section, puis on agrège les estimés
    - Permet de trier par déciles mensuels (comme dans la Table 1)

- **Gérer naturellement les « leads » et « lags »**
    - Pour mesurer l’impact de βUNC sur le rendement futur (leads) ou passé (lags)

## Utilité dans notre contexte

- **Déciles de βUNC** : à chaque date, on trie toutes les actions et on calcule le rendement moyen du décile 1 à 10 → données de panel pures.
- **Régressions factorielles par décile** : pour chaque décile, on récolte une série temporelle des excès-rendements, que l’on régresse sur les facteurs (CAPM, 5F, 7F) avec Newey–West.
- **Table 2 (Fama–MacBeth)** : on construit un panel où, pour chaque titre et chaque mois, on a βUNC et d’autres firm-level characteristics (ILLIQ, COSKEW, IVOL, MAX, SIZE, BM, …), puis on effectue une série de cross-sections par date, avant d’agréger les coefficients.

> **En résumé :**  
> Le panel est le format de travail de base : un grand tibble « date × ticker » dans lequel on peut trier, regrouper, regresser et créer toutes les variables leads/lags nécessaires.  
> Cela permet de réconcilier les effets spécifiques aux titres (cross-section) et les dynamiques temporelles (séries temporelles) dans une même analyse.

```r
library(dplyr)
library(purrr)
library(broom)
library(sandwich)
library(lmtest)

# ──────────────────────────────────────────────────────────────────────────────
# 1) Fonction utilitaire : Newey–West sur la série des betas cross-sectionnels
# ──────────────────────────────────────────────────────────────────────────────
fm_beta <- function(horizon, controls = NULL, panel_df, nw_lag = 6) {
  lead_col <- paste0("beta_unc_lead", horizon)
  # (a) Régressions cross-sectionnelles mois par mois
  cs <- panel_df %>%
    filter(!is.na(.data[[lead_col]])) %>%
    group_by(date) %>%
    do(tidy(
      lm(
        as.formula(
          paste0(lead_col, " ~ beta_unc",
                 if(length(controls)) paste0(" + ", paste(controls, collapse=" + ")) else "")
        ),
        data = .
      )
    )) %>%
    ungroup() %>%
    filter(term == "beta_unc")
  # (b) Fama–MacBeth : moyenne des estimateurs + t-stat Newey–West
  m  <- lm(estimate ~ 1, data = cs)  # moyenne des bêtas cross-sectionnels
  V  <- NeweyWest(m, lag = nw_lag, prewhite = FALSE)  # correction de l'autocorrélation
  b0 <- coef(m)[1]
  t0 <- b0 / sqrt(V[1,1])
  tibble(
    horizon      = paste0(horizon, " an(s)"),
    beta         = b0,
    t_nw         = t0
  )
}
```

```r

# ──────────────────────────────────────────────────────────────────────────────
# 2) Liste des contrôles firm‐level pour Table 2
# ──────────────────────────────────────────────────────────────────────────────
controls <- c("ILLIQ", "COSKEW", "IVOL", "MAX")
```

```r
# ──────────────────────────────────────────────────────────────────────────────
# (3) Construction du panel avec lag/lead et firm-level variables
# ──────────────────────────────────────────────────────────────────────────────
library(dplyr)

panel_cleaned <- uncertainty_betas %>%
  arrange(ticker, date) %>%
  group_by(ticker) %>%
  mutate(
    beta_unc_lag   = lag(beta_unc,   1),     # βUNC au mois précédent (t-1)
    beta_unc_lead1 = lead(beta_unc, 12),     # βUNC à t+12 mois
    beta_unc_lead2 = lead(beta_unc, 24),     # βUNC à t+24 mois
    beta_unc_lead3 = lead(beta_unc, 36),     # βUNC à t+36 mois
    beta_unc_lead4 = lead(beta_unc, 48),     # βUNC à t+48 mois
    beta_unc_lead5 = lead(beta_unc, 60)      # βUNC à t+60 mois
  ) %>%
  ungroup() %>%
  # Jointures avec les variables firm-level préparées
  left_join(ILLIQ_df,  by = c("date","ticker")) %>%
  left_join(COSKEW_df, by = c("date","ticker")) %>%
  left_join(IVOL_df,   by = c("date","ticker")) %>%
  left_join(MAX_df,    by = c("date","ticker")) %>%
  left_join(
    Q_factors %>% select(date, R_MKT, R_IA, R_ROE),
    by = "date"
  ) %>%
  # On ne conserve que les observations pour lesquelles on a au moins le lag et le lead principal
  filter(
    !is.na(beta_unc_lag),
    !is.na(beta_unc_lead1)
    # (on laisse les NA sur les contrôles, qui seront gérés par modèle)
  )
```

> **Pourquoi de nombreuses valeurs de `beta_unc` sont NA au début de la série ?**
>
> Lors de la construction du panel, on filtre généralement les lignes pour ne garder que celles où `beta_unc` n’est **pas NA**.
>
> Cela s’explique par la méthode de calcul de `beta_unc` :  
> - **`beta_unc` est estimé via une régression sur une fenêtre mobile de 60 mois.**  
> - On utilise donc `rollapply` avec `width = 60` (c’est-à-dire une fenêtre de 60 observations mensuelles consécutives).
>
> **Conséquence :**  
> - Pour chaque titre, il est impossible de calculer un beta sur les 59 premiers mois (la fenêtre n’est pas complète).  
> - Le tout **premier `beta_unc` non-NA** pour une action apparaît donc au **60e mois** après le début de la série.
> - Par exemple, si la série commence en janvier 1999, il faut attendre janvier 2004 pour obtenir le premier beta estimé.
>
> **Remarque** :  
> - Ce comportement est standard pour tout estimateur « rolling » : les toutes premières dates sont perdues car il n’y a pas assez d’historique.
> - Ce filtrage explique pourquoi, lors de l’analyse ou du merge, la période réellement exploitable débute plus tardivement que la période brute disponible dans les données.

```r
# ──────────────────────────────────────────────────────────────────────────────
# 4) Boucle sur les horizons 1 à 5 ans pour Table 2
# ──────────────────────────────────────────────────────────────────────────────

table2 <- map_dfr(1:5, function(h) {
  # Régression univariée (sans contrôles)
  uni  <- fm_beta(h, controls = NULL,     panel_df = panel_cleaned)
  # Régression multivariée (avec contrôles)
  mult <- fm_beta(h, controls = controls, panel_df = panel_cleaned)
  # Mise en forme du résultat
  uni  %>%
    rename(uni_beta   = beta,
           uni_tstat  = t_nw) %>%
    left_join(
      mult %>% rename(ctrl_beta  = beta,
                      ctrl_tstat = t_nw),
      by = "horizon"
    )
})

# Affichage du résultat
print(table2)
```

## Table 3 — Déterminants du Bêta d’Incertitude : Analyse Cross-Sectionnelle Mensuelle

La Table 3 présente une analyse approfondie des déterminants du bêta d’incertitude (**βUNC**) à l’échelle individuelle des actions. Cette table est construite à partir de régressions cross-sectionnelles réalisées chaque mois, où l’on cherche à expliquer le **βUNC futur à 12 mois** par différentes caractéristiques financières et comportementales propres à chaque entreprise.

Les variables explicatives intégrées dans ces régressions sont issues de la littérature financière et couvrent un large spectre des dimensions de risque, de valorisation et de comportement des investisseurs :

- **Taille (SIZE)** : mesure la capitalisation boursière de l’entreprise, associée à la liquidité et à la stabilité du titre.
- **Book-to-Market (BM)** : ratio valeur comptable/valeur de marché, indicateur classique du style « value ».
- **Reversal (REV)** : rendement du mois précédent, traduisant la tendance au retour à la moyenne à court terme.
- **Momentum (MOM)** : performance passée sur une période moyenne, reflétant la persistance des rendements.
- **Illiquidité (ILLIQ)** : capacité à échanger le titre sans impacter significativement son prix, proxy du risque de liquidité.
- **Co-skewness (COSKEW)** : sensibilité du rendement du titre à la convexité des rendements du marché, capturant l’exposition aux événements extrêmes.
- **Volatilité idiosyncratique (IVOL)** : composante du risque propre à l’entreprise, non expliquée par les facteurs de marché.
- **Dispersion des prévisions d’analystes (DISP)** : mesure le désaccord entre analystes, proxy de l’incertitude perçue par le marché.
- **Bêta de volatilité de marché (βVXO)** : sensibilité du titre aux variations de la volatilité globale du marché.
- **Investissement (I/A)** : taux d’investissement de l’entreprise, indicateur de sa politique d’expansion.
- **Rentabilité (ROE)** : Return on Equity, reflet de la performance opérationnelle.
- **Lottery demand (MAX)** : propension du titre à réaliser des rendements extrêmes, caractéristique des actions prisées pour leur potentiel de « gains loterie ».

Chaque mois, ces variables sont utilisées pour expliquer la variation cross-sectionnelle du **bêta d’incertitude à venir** via des régressions, individuellement ou en combinaison. Les coefficients moyens reportés dans la table révèlent les liens structurels et robustes entre ces caractéristiques de firme et leur sensibilité à l’incertitude macroéconomique, offrant ainsi une lecture précieuse pour la gestion du risque et la construction de portefeuilles en environnement incertain.

```r
# ──────────────────────────────────────────────────────────────────────────────
# 2) Fonction utilitaire Newey-West, régression et table
# ──────────────────────────────────────────────────────────────────────────────

library(dplyr)
library(purrr)
library(broom)
library(sandwich)

# Fonction utilitaire : calcule l'alpha et le t-stat Newey–West sur une série d'estimateurs
nw_coef <- function(cs, lag = 6) {
  # cs doit être un tibble à une colonne "estimate"
  fit <- lm(estimate ~ 1, data = cs)
  V   <- NeweyWest(fit, lag = lag, prewhite = FALSE)
  beta <- coef(fit)[1]
  t    <- beta / sqrt(V[1,1])
  list(beta = beta, t = t)
}

# Nos variables univariées à tester
controls_uni <- c("beta_unc_lag", "ILLIQ", "COSKEW",
                  "IVOL",        "MAX")

# Boucle robuste qui ne plante jamais
uni_results <- map_dfr(controls_uni, function(ctrl) {
  # 1) on filtre d’abord sur lead1 et sur le contrôle courant
  df0 <- panel %>%
    filter(!is.na(beta_unc_lead1), !is.na(.data[[ctrl]]))

  # 2) on ne conserve que les dates où il y a >=2 obs
  valid <- df0 %>%
    count(date) %>%
    filter(n >= 2) %>%
    pull(date)

  # 3) si aucune date valide, on renvoie un seul tibble d’NA
  if (length(valid) == 0) {
    return(tibble(
      variable = ctrl,
      uni_beta = NA_real_,
      uni_t    = NA_real_
    ))
  }

  # 4) sinon on construit la série cross-sectionnelle des betas
  cs <- df0 %>%
    filter(date %in% valid) %>%
    group_by(date) %>%
    # tryCatch pour intercepter d’éventuelles erreurs malgré tout
    do({
      fit <- tryCatch(
        lm(as.formula(paste0("beta_unc_lead1 ~ ", ctrl)), data = .),
        error = function(e) NULL
      )
      if (is.null(fit)) {
        tibble(estimate = NA_real_)
      } else {
        tidy(fit) %>%
          filter(term == ctrl) %>%
          select(estimate)
      }
    }) %>%
    ungroup()

  # 5) si tidy() a renvoyé zéro ligne (impossible en principe), on met NA
  if (nrow(cs) == 0) {
    return(tibble(
      variable = ctrl,
      uni_beta = NA_real_,
      uni_t    = NA_real_
    ))
  }

  # 6) enfin on calcule la moyenne et le t-stat NW
  stats <- nw_coef(cs, lag = 6)
  tibble(
    variable = ctrl,
    uni_beta = stats$beta,
    uni_t    = stats$t
  )
})

uni_results
```

## Régressions cross-sectionnelles mensuelles et agrégation temporelle — Table 3

### Régression mensuelle

Pour chaque mois t, nous estimons la régression cross-sectionnelle suivante :

    beta_unc_{i,t} = lambda_{0,t} + lambda_{1,t} * X_{i,t} + epsilon_{i,t}

où X_{i,t} représente soit une seule caractéristique de firme (spécification univariée), soit l’ensemble des contrôles simultanément (spécification multivariée).

---

### Agrégation temporelle

Après estimation, nous obtenons pour chaque mois une série de coefficients {lambda_hat_{1,t}}. Nous calculons ensuite la moyenne temporelle :

    mean_lambda_1 = (1/T) * sum_{t=1}^{T} lambda_hat_{1,t}

ainsi que son t-stat, en régressant lambda_hat_{1,t} sur une constante, avec correction Newey–West pour l’autocorrélation et l’hétéroscédasticité.

---

### Spécifications retenues

- **Colonnes (1) à (13) de la Table 3** : régressions univariées, où X_{i,t} prend tour à tour la valeur de chaque caractéristique (SIZE, BM, REV, MOM, ILLIQ, COSKEW, IVOL, DISP, beta_VXO, I/A, ROE, MAX, etc.).
- **Colonne (14)** : régression multivariée où toutes les caractéristiques sont incluses simultanément dans X.

### Implémentation

Nous avons procédé en suivant strictement ces étapes :

- Boucle sur chaque caractéristique retenue ;
- Pour chaque mois : estimation d’une régression cross-sectionnelle (lm) entre beta_unc_{i,t} et la/les caractéristiques ;
- Stockage des coefficients lambda_hat_{1,t} ;
- Calcul de la moyenne et du t-statistique Newey–West sur la série temporelle des coefficients.

## Comment passer du panel de données à la Table 4 (Panel A)

Pour construire la Table 4 (Panel A), nous procédons ainsi :

1. **Régressions cross-sectionnelles mensuelles**

   Pour chaque mois $t$, nous estimons la régression suivante :

       exret_{i,t} = α_t
                     + β^{UNC}_t × beta_{UNC,i,t-60:t-1}
                     + β^{MKT}_t × beta_{MKT,i,t-60:t-1}
                     + β^{VXO}_t × beta_{VXO,i,t-60:t-1}
                     + γ_1 × SIZE_{i,t}
                     + γ_2 × BM_{i,t}
                     + γ_3 × MOM_{i,t}
                     + ...
                     + ε_{i,t}

   où chaque variable correspond à la valeur moyenne ou caractéristique sur la fenêtre de 60 mois précédents (ou à la date $t$ pour les firm-traits).  
   On inclut toutes les variables listées en colonnes de la Table 4 (SIZE, BM, MOM, REV, ILLIQ, COSKEW, IVOL, DISP, I/A, ROE, MAX), et éventuellement la variable industrie si l’on souhaite contrôler l’effet industrie.

2. **Collecte des coefficients**

   Nous collectons, pour chaque mois, les coefficients estimés pour chaque variable :  
   - β^{UNC}_t, β^{MKT}_t, β^{VXO}_t, γ_1, γ_2, ..., etc.

3. **Agrégation temporelle**

   Pour chaque variable, nous calculons :
   - La moyenne temporelle du coefficient estimé (moyenne sur tous les mois).
   - Le t-statistique associé, en régressant la série mensuelle de coefficients sur une constante, avec correction Newey–West pour l’autocorrélation et l’hétéroscédasticité.

4. **Présentation des résultats**

   - On reporte pour chaque variable la moyenne des coefficients et le t-stat corrigé Newey–West.
   - Les colonnes (1) à (6) de Panel A sont obtenues sans effet industrie, les colonnes (7) à (12) avec l’effet industrie.

---

**Résumé du workflow :**

- Pour chaque mois, estimation d’une régression cross-sectionnelle avec toutes les variables de la Table 4.
- Stockage mensuel des coefficients estimés.
- Calcul, pour chaque variable, de la moyenne et du t-stat Newey–West sur la période.
- Présentation dans la Table 4 (Panel A).

---

**Remarque :**  
Si certaines variables (par exemple β_{VXO}, REV, DISP ou les effets industrie) ne sont pas disponibles dans notre panel, nous ne pouvons pas remplir toutes les colonnes de la Table 4, mais il reste possible de produire les colonnes correspondant aux variables effectivement présentes.

```r
library(dplyr)
library(broom)
library(sandwich)
library(lmtest)

fm_monthly <- function(formula_str, panel, lags = 6) {
  # 1) Transformez la chaîne en formule
  f <- as.formula(formula_str)
  # 2) Récupérez tous les noms de variables de la formule
  vars <- all.vars(f)
  # 3) Ne gardez que les observations sans NA sur ces variables
  panel_clean <- panel %>%
    select(date, all_of(vars)) %>%
    drop_na()

  # 4) Cross‐section par date
  cs <- panel_clean %>%
    group_by(date) %>%
    do(tidy(lm(f, data = .))) %>%
    ungroup() %>%
    filter(term == "beta_unc")

  # 5) Fama–MacBeth : moyenne et t‐stat NW
  m0 <- lm(estimate ~ 1, data = cs)
  V0 <- NeweyWest(m0, lag = lags, prewhite = FALSE)
  est <- coef(m0)[1]
  t0  <- est / sqrt(V0[1,1])

  list(alpha = est, t_stat = t0)
}
```

```r
library(dplyr)
library(broom)
library(sandwich)
library(lmtest)

# Fonction pour estimer des régressions Fama–MacBeth mensuelles sur l'ensemble des variables d'une formule
fm_monthly_all_terms <- function(formula_str, panel, lags = 6) {
  f    <- as.formula(formula_str)           # Conversion de la chaîne en objet formule R
  vars <- all.vars(f)                       # Extraction de toutes les variables de la formule
  panel_clean <- panel %>%
    select(date, all_of(vars)) %>%          # Sélection des colonnes utiles (date + variables de la formule)
    drop_na()                               # Suppression des lignes incomplètes (NA)

  # Pour chaque mois, estimation d'une régression cross-sectionnelle et extraction des coefficients (tidy)
  cs <- panel_clean %>%
    group_by(date) %>%
    do(tidy(lm(f, data = .))) %>%           # tidy() permet de garder tous les coefficients de la régression
    ungroup()

  # Moyenne et t-stat Newey-West par variable (terme de la régression)
  cs %>%
    group_by(term) %>%
    summarize(
      estimate = mean(estimate, na.rm = TRUE), # Moyenne temporelle du coefficient
      t_stat   = {
        e <- estimate
        if(length(e) > 1) {
          m0 <- lm(estimate ~ 1, data = data.frame(estimate = e)) # Régression sur constante
          V0 <- NeweyWest(m0, lag = lags, prewhite = FALSE)        # Correction Newey–West
          coef(m0)[1] / sqrt(V0[1,1])                             # Calcul du t-stat corrigé
        } else NA_real_
      },
      .groups = "drop"
    )
}

# Exemple d'utilisation : estimation de la spécification 4 avec toutes les variables correspondantes
table_terms_4 <- fm_monthly_all_terms(
  "exret ~ beta_unc + beta_mkt + LT_REV + ILLIQ",
  panel_T4
)
print(table_terms_4)
```

```r
# ──────────────────────────────────────────────────────────────────────────────
# 1) Packages
# ──────────────────────────────────────────────────────────────────────────────
library(dplyr); library(tidyr); library(readxl)
library(zoo);    library(sandwich); library(lmtest)
library(purrr)

# ──────────────────────────────────────────────────────────────────────────────
# 2) Mettre Risk_Premium en format long + calcul de l'excès de rendement
# ──────────────────────────────────────────────────────────────────────────────
rp_long <- Risk_Premium %>%
  mutate(across(-date, as.numeric)) %>% # Conversion des rendements en numérique
  mutate(date = as.Date(date)) %>%
  pivot_longer(-date, names_to="ticker", values_to="raw_ret") %>% # Passage au format long (ticker, date)
  left_join(
    Q_factors %>% select(date, R_F) %>% mutate(R_F = as.numeric(R_F)),
    by="date"
  ) %>%
  mutate(exret = raw_ret - R_F) %>% # Calcul de l'excès de rendement sur l'actif sans risque
  filter(!is.na(exret))

# ──────────────────────────────────────────────────────────────────────────────
# 3) Préparer l'indicateur UNC_{3m} décalé d’un mois (pour anticiper l'incertitude future)
# ──────────────────────────────────────────────────────────────────────────────
unc3_next <- MacroUncertainty %>%
  mutate(date = as.Date(Date)) %>%
  arrange(date) %>%
  transmute(
    date,
    UNC3_next = lead(`UNC_3m`, 1) # Décalage d'un mois vers l'avenir
  ) %>%
  filter(!is.na(UNC3_next))

# ──────────────────────────────────────────────────────────────────────────────
# 4) Calculer β^UNC_{3m} en fenêtre glissante de 60 mois pour chaque titre
# ──────────────────────────────────────────────────────────────────────────────
betas3 <- rp_long %>%
  inner_join(unc3_next, by="date") %>%
  arrange(ticker, date) %>%
  group_by(ticker) %>%
  mutate(
    beta_unc3 = rollapplyr(
      data      = cbind(exret, UNC3_next),
      width     = 60,
      by.column = FALSE,
      fill      = NA_real_,
      align     = "right",
      FUN       = function(z) {
        if(any(is.na(z))) return(NA_real_) # Ignore les fenêtres incomplètes
        coef(lm(z[,1] ~ z[,2]))[2] # Slope de la régression locale
      }
    )
  ) %>%
  ungroup() %>%
  select(date, ticker, beta_unc3) %>%
  filter(!is.na(beta_unc3)) # On ne garde que les fenêtres valides

# ──────────────────────────────────────────────────────────────────────────────
# 5) Former les portefeuilles décile‐par‐décile selon β^UNC_{3m}
# ──────────────────────────────────────────────────────────────────────────────
portfolio3 <- betas3 %>%
  inner_join(rp_long, by=c("date","ticker")) %>%
  group_by(date) %>%
  mutate(decile = ntile(beta_unc3, 10)) %>% # Attribution du décile à chaque titre
  ungroup() %>%
  group_by(date, decile) %>%
  summarise(
    avg_beta   = mean(beta_unc3, na.rm=TRUE), # Moyenne des βUNC par décile
    avg_return = mean(raw_ret,  na.rm=TRUE),  # Rendement brut moyen
    exret      = mean(exret,    na.rm=TRUE),  # Excès de rendement moyen
    .groups    = "drop"
  )

# ──────────────────────────────────────────────────────────────────────────────
# 6) Fonction d’aide : calcul alpha & t-stat intercept avec Newey–West
# ──────────────────────────────────────────────────────────────────────────────
safe_nw <- function(m, lag=12){
  tryCatch({
    V   <- NeweyWest(m, lag=lag, prewhite=FALSE)
    ct  <- coeftest(m, vcov=V)
    list(alpha = ct[1,"Estimate"], t_stat = ct[1,"t value"])
  }, error = function(e){
    list(alpha=NA_real_, t_stat=NA_real_)
  })
}

# ──────────────────────────────────────────────────────────────────────────────
# 7) Calcul des alphas (intercept) décile‐par‐décile sous différents modèles
#     (CAPM, 5f_JLN, 5f_HXZ, 4f_HXZ, 7f)
# ──────────────────────────────────────────────────────────────────────────────
alpha_h3 <- portfolio3 %>%
  group_by(decile) %>%
  filter(n() >= 24) %>% # On garde les déciles avec un historique suffisant
  group_map(~{
    df <- .x %>%
      left_join(Q_factors2, by="date") %>%
      drop_na(exret, R_MKT, R_ME, HML, MOM, LIQ, R_IA, R_ROE)

    # Régressions pour chaque modèle factoriel
    m1  <- lm(exret ~ R_MKT,                             data=df); nw1  <- safe_nw(m1)
    m5a <- lm(exret ~ R_MKT+R_ME+HML+MOM+LIQ,           data=df); nw5a <- safe_nw(m5a)
    m5b <- lm(exret ~ R_MKT+R_ME+HML+R_IA+R_ROE,          data=df); nw5b <- safe_nw(m5b)
    m4  <- lm(exret ~ R_MKT+R_ME+R_IA+R_ROE,              data=df); nw4  <- safe_nw(m4)
    m7  <- lm(exret ~ R_MKT+R_ME+HML+MOM+LIQ+R_IA+R_ROE, data=df); nw7  <- safe_nw(m7)

    # Résumé pour chaque décile
    tibble(
      decile        = as.character(.y$decile),
      avg_beta      = mean(.x$avg_beta),
      avg_return    = mean(.x$avg_return),
      exret         = mean(.x$exret),
      capm_alpha    = nw1$alpha,   capm_t    = nw1$t_stat,
      alpha_5f_JLN  = nw5a$alpha,  t_5f_JLN  = nw5a$t_stat,
      alpha_5f_HXZ  = nw5b$alpha,  t_5f_HXZ  = nw5b$t_stat,
      alpha_4f_HXZ  = nw4$alpha,   t_4f_HXZ  = nw4$t_stat,
      alpha_7f      = nw7$alpha,   t_7f      = nw7$t_stat
    )
  }) %>%
  bind_rows() %>%
  arrange(as.integer(decile))

# ──────────────────────────────────────────────────────────────────────────────
# 8) Calcul du spread “10-1” (décile extrême haut - extrême bas)
# ──────────────────────────────────────────────────────────────────────────────
spread3 <- alpha_h3 %>%
  filter(decile %in% c("1","10")) %>%
  summarise(
    decile         = "10-1",
    avg_beta       = diff(avg_beta),
    avg_return     = diff(avg_return),
    exret          = diff(exret),
    capm_alpha     = diff(capm_alpha),    capm_t     = NA_real_,
    alpha_5f_JLN   = diff(alpha_5f_JLN),  t_5f_JLN   = NA_real_,
    alpha_5f_HXZ   = diff(alpha_5f_HXZ),  t_5f_HXZ   = NA_real_,
    alpha_4f_HXZ   = diff(alpha_4f_HXZ),  t_4f_HXZ   = NA_real_,
    alpha_7f       = diff(alpha_7f),      t_7f       = NA_real_
  )

alpha_results_h3 <- bind_rows(alpha_h3, spread3)

# ──────────────────────────────────────────────────────────────────────────────
# 9) Affichage final : résultats Table 5 “approximation décile”
# ──────────────────────────────────────────────────────────────────────────────
print(alpha_results_h3)
```

```r
# ──────────────────────────────────────────────────────────────────────────────
# 1) Packages
# ──────────────────────────────────────────────────────────────────────────────
library(dplyr)
library(tidyr)
library(readxl)
library(zoo)
library(sandwich)
library(lmtest)
library(purrr)

# ──────────────────────────────────────────────────────────────────────────────
# 2) Risk_Premium en long + rendement excédentaire (R_F déjà dans Q_factors)
# ──────────────────────────────────────────────────────────────────────────────
rp_long <- Risk_Premium %>%
  mutate(across(-date, as.numeric)) %>%
  mutate(date = as.Date(date)) %>%
  pivot_longer(-date, names_to="ticker", values_to="raw_ret") %>%
  left_join(
    Q_factors %>%
      select(date, R_F) %>%
      mutate(R_F = as.numeric(R_F)),
    by="date"
  ) %>%
  mutate(exret = raw_ret - R_F) %>%
  filter(!is.na(exret))

# ──────────────────────────────────────────────────────────────────────────────
# 3) Préparer UNC_{12m} décalé d’un mois
# ──────────────────────────────────────────────────────────────────────────────
unc12_next <- MacroUncertainty %>%
  rename(date = Date) %>%            # si votre colonne s’appelle "Date"
  mutate(date = as.Date(date)) %>%
  arrange(date) %>%
  transmute(
    date,
    UNC12_next = lead(UNC_12m, 1)
  ) %>%
  filter(!is.na(UNC12_next))

# ──────────────────────────────────────────────────────────────────────────────
# 4) Calculer β^UNC_{12m} en rolling window de 60 mois
# ──────────────────────────────────────────────────────────────────────────────
betas12 <- rp_long %>%
  inner_join(unc12_next, by="date") %>%
  arrange(ticker, date) %>%
  group_by(ticker) %>%
  mutate(
    beta_unc12 = rollapplyr(
      data      = cbind(exret, UNC12_next),
      width     = 60,
      by.column = FALSE,
      fill      = NA_real_,
      align     = "right",
      FUN       = function(z) {
        if(any(is.na(z))) return(NA_real_)
        coef(lm(z[,1] ~ z[,2]))[2]
      }
    )
  ) %>%
  ungroup() %>%
  select(date, ticker, beta_unc12) %>%
  filter(!is.na(beta_unc12))

# ──────────────────────────────────────────────────────────────────────────────
# 5) Former les portefeuilles égal‐pondérés décile‐par‐décile
# ──────────────────────────────────────────────────────────────────────────────
portfolio12 <- betas12 %>%
  inner_join(rp_long, by = c("date","ticker")) %>%
  group_by(date) %>%
  mutate(decile = ntile(beta_unc12, 10)) %>%
  ungroup() %>%
  group_by(date, decile) %>%
  summarise(
    avg_beta   = mean(beta_unc12, na.rm = TRUE),
    avg_return = mean(raw_ret,     na.rm = TRUE),
    exret      = mean(exret,       na.rm = TRUE),
    .groups    = "drop"
  )

# ──────────────────────────────────────────────────────────────────────────────
# 6) Helper : Newey–West sur l’intercept (safe)
# ──────────────────────────────────────────────────────────────────────────────
safe_nw <- function(mod, lag = 12) {
  tryCatch({
    V  <- NeweyWest(mod, lag = lag, prewhite = FALSE)
    ct <- coeftest(mod, vcov = V)
    list(alpha  = ct[1,"Estimate"],
         t_stat = ct[1,"t value"])
  }, error = function(e) {
    list(alpha  = NA_real_, t_stat = NA_real_)
  })
}

# ──────────────────────────────────────────────────────────────────────────────
# 7) Calcul des alphas décile‐par‐décile (CAPM, 5f_JLN, 5f_HXZ, 4f_HXZ, 7f)
# ──────────────────────────────────────────────────────────────────────────────
alpha_h12 <- portfolio12 %>%
  group_by(decile) %>%
  filter(n() >= 24) %>%
  group_map(~{
    df <- .x %>%
      left_join(Q_factors2, by = "date") %>%
      drop_na(
        exret, R_MKT, R_ME, HML, MOM,
        LIQ,    # votre factor de liquidité agrégée
        R_IA, R_ROE
      )

    # Modèles
    m1  <- lm(exret ~ R_MKT,                             data = df); nw1  <- safe_nw(m1)
    m5a <- lm(exret ~ R_MKT + R_ME + HML + MOM + LIQ,  data = df); nw5a <- safe_nw(m5a)
    m5b <- lm(exret ~ R_MKT + R_ME + HML + R_IA + R_ROE,  data = df); nw5b <- safe_nw(m5b)
    m4  <- lm(exret ~ R_MKT + R_ME + R_IA + R_ROE,        data = df); nw4  <- safe_nw(m4)
    m7  <- lm(exret ~ R_MKT + R_ME + HML + MOM + LIQ + R_IA + R_ROE,
                data = df); nw7  <- safe_nw(m7)

    tibble(
      decile        = as.character(.y$decile),
      avg_beta      = mean(.x$avg_beta,   na.rm = TRUE),
      avg_return    = mean(.x$avg_return, na.rm = TRUE),
      exret         = mean(.x$exret,      na.rm = TRUE),

      capm_alpha    = nw1$alpha,   capm_t    = nw1$t_stat,
      alpha_5f_JLN  = nw5a$alpha,  t_5f_JLN  = nw5a$t_stat,
      alpha_5f_HXZ  = nw5b$alpha,  t_5f_HXZ  = nw5b$t_stat,
      alpha_4f_HXZ  = nw4$alpha,   t_4f_HXZ  = nw4$t_stat,
      alpha_7f      = nw7$alpha,   t_7f      = nw7$t_stat
    )
  }) %>%
  bind_rows() %>%
  arrange(as.integer(decile))

# ──────────────────────────────────────────────────────────────────────────────
# 8) Ajouter la ligne “High–Low” (10–1)
# ──────────────────────────────────────────────────────────────────────────────
spread12 <- alpha_h12 %>%
  filter(decile %in% c("1","10")) %>%
  summarise(
    decile        = "10-1",
    avg_beta      = diff(avg_beta),
    avg_return    = diff(avg_return),
    exret         = diff(exret),
    capm_alpha    = diff(capm_alpha),   capm_t    = NA_real_,
    alpha_5f_JLN  = diff(alpha_5f_JLN), t_5f_JLN = NA_real_,
    alpha_5f_HXZ  = diff(alpha_5f_HXZ), t_5f_HXZ = NA_real_,
    alpha_4f_HXZ  = diff(alpha_4f_HXZ), t_4f_HXZ = NA_real_,
    alpha_7f      = diff(alpha_7f),     t_7f     = NA_real_
  )

alpha_results_h12 <- bind_rows(alpha_h12, spread12)

# ──────────────────────────────────────────────────────────────────────────────
# 9) Affichage final : Table 6 (h = 12 mois)
# ──────────────────────────────────────────────────────────────────────────────
print(alpha_results_h12)
```

  Certaines industries du S&P 500 ne comptent que 10 à 15 actions. Il devient alors impossible de découper chaque industrie en 10 groupes (déciles) de manière statistiquement significative ou stable.

- **Exigence d’un minimum de 24 mois d’historique** :  
  Pour estimer un alpha Newey–West robuste, il faut au moins 24 mois d’observations par groupe, comme le retient l’article.  
  Avec un échantillon restreint, seuls les déciles les plus larges (typiquement le bas et le milieu de la distribution) parviennent à satisfaire ce critère, tandis que les extrêmes (déciles 1 et 10) peuvent manquer de données ou être trop étroits.

## Pourquoi la Table 7 n’affiche que 5 déciles par industrie sur S&P 500

Notre Table 7 actuelle (50 lignes, soit 5 déciles sur 10 par industrie) est la seule possible lorsqu’on respecte la méthode et le critère des 24 mois sur un échantillon limité au S&P 500.

### Pourquoi n’obtient-on pas 10 déciles complets ?

- **Taille de l’univers :**  
  L’univers S&P 500, une fois toutes les contraintes appliquées, ne compte que 258 titres exploitables (actions avec données complètes).  
  Or, certaines industries n’incluent qu’un nombre très limité d’actions.

- **Critère d’ancienneté :**  
  Pour chaque décile, on exige au moins 24 observations mensuelles (filter(n() >= 24)) afin de garantir la robustesse des estimations (en particulier pour l’alpha Newey–West).

- **Conséquence directe :**  
  Il devient impossible de constituer 10 déciles par industrie avec au moins 24 mois dans chaque décile.  
  En pratique, seules les 5 premières déciles (1 à 5) disposent de suffisamment d’observations pour être conservées, tandis que les déciles 6 à 10 sont filtrés et disparaissent de la Table 7.

### Comment obtenir les 10 déciles ?

- **Élargir l’univers :**  
  Il faudrait passer à un univers plus large, par exemple CRSP (toutes les actions US), pour disposer d’assez de titres par industrie et par décile.
- **Ou relâcher le critère d’historique :**  
  Accepter des déciles très courts (moins de 24 mois), mais ceci dégraderait la robustesse des résultats.

---

**En résumé :**  
L’univers S&P 500 réduit et la contrainte d’un minimum de 24 mois expliquent pourquoi seules les 5 premières déciles par industrie apparaissent dans notre Table 7.
